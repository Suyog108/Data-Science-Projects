```python
import pandas as pd #Data Manupulation
import numpy as np #numerical operations
from sklearn import svm, datasets #for SVM classification
import matplotlib.pyplot as plt #data visualization
```

```python
iris = datasets.load_iris()
```

```python
X = iris.data[:, :2]
y = iris.target
```

```python
#following values define the boundaries of the plot based on mini and max values of sepal length 
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
#h is small step size used to create mesh grid for plotting decision boundaries
h = (x_max / x_min)/100
#xx and yy create a mesh grid using numpy's meshgrid function
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
   np.arange(y_min, y_max, h))
X_plot = np.c_[xx.ravel(), yy.ravel()]
```

```python
C = 1.0
```

```python
#create svm model with linear kernal
svc_classifier = svm.SVC(kernel='linear', C=C).fit(X, y)#train svm kernel on data
Z = svc_classifier.predict(X_plot) #predict class label
Z = Z.reshape(xx.shape) #z reshaped to match the shape of 'xx' for plotting
plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.contourf(xx, yy, Z, cmap=plt.cm.tab10, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('Support Vector Classifier with linear kernel')
```

```python
#create svm model with rbf kernal
Svc_classifier = svm.SVC(kernel='rbf', gamma ='auto',C=C).fit(X, y)
Z = svc_classifier.predict(X_plot)
Z = Z.reshape(xx.shape)
plt.figure(figsize=(15, 5))
plt.subplot(121)
plt.contourf(xx, yy, Z, cmap=plt.cm.tab10, alpha=0.3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.xlim(xx.min(), xx.max())
plt.title('Support Vector Classifier with rbf kernel')
```
