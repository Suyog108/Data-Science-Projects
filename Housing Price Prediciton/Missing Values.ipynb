```python
#Treating Missing values
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
X_full = pd.read_csv(r"D:\Housing price\train.csv")
X_test_full = pd.read_csv(r"D:\Housing price\test.csv")
```

```python
X_full.shape
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows',None)
```

```python
X_full.head()
```

```python
y = X_full.SalePrice
```

```python
features = ['LotArea','LotFrontage', 'MSSubClass','OverallQual','OverallCond','BsmtFinSF1','BsmtUnfSF', 'TotalBsmtSF', 'GrLivArea', 'GarageArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = X_full[features].copy()
X_test = X_test_full[features].copy()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
X_train.head()
```

```python
X_train.shape
```

```python
#the 0 in the data is replaced by NAN to tryout which Missing value methords fits best
```

```python
X_train.isnull().sum()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
def score_dataset (X_train, X_valid, y_trian, y_valid):
    model = RandomForestRegressor(n_estimators=100, criterion='absolute_error', random_state=0)
    model.fit(X_train, y_train)
    preds = model.predict(X_valid)
    return mean_absolute_error(y_valid, preds)
```

```python
#Approach 1 to Drop the columns with Missing values
#Get names of column with missing values
cols_with_missing = [col for col in X_train.columns
                     if X_train[col].isnull().any()]
```

```python
print(cols_with_missing)
```

```python
#To drop to entire column if there are any null values in the column
reduced_X_train = X_train.drop(cols_with_missing, axis=1)
reduced_X_valid = X_valid.drop(cols_with_missing, axis=1)
```

```python
reduced_X_train.shape
```

```python
MAE1 = score_dataset(reduced_X_train, reduced_X_valid, y_train, y_valid)
print("MAE from Approach 1 reduce column: %d" % MAE1)
```

```python
#The MAE is reduced in Approach 1 due to the reduction in column we lost valuable data which compounded to higher accuracy
```

```python
#Approach 2 to replace missing value with mean value.
from sklearn.impute import SimpleImputer
```

```python
my_imputer = SimpleImputer()
imputed_X_train = pd.DataFrame(my_imputer.fit_transform(X_train))
imputed_X_valid = pd.DataFrame(my_imputer.transform(X_valid))
#Imputation removed column names put them back
imputed_X_train.columns = X_train.columns
imputed_X_valid.columns = X_valid.columns
```

```python
imputed_X_train.head()
```

```python
MAE2 = score_dataset(imputed_X_train, imputed_X_valid, y_train, y_valid)
print("MAE from Approach 2 Imputation:%d" % MAE2)
```

```python
#Approach 3 Impute missing value while keeping track of values which were impute
```

```python
X_train_plus = X_train.copy()
X_valid_plus = X_valid.copy()
```

```python
#Make new columns indicating what will be imputed
for col in cols_with_missing:
    X_train_plus[col+'_was_missing'] = X_train_plus[col].isnull()
    X_valid_plus[col+'_was_missing'] = X_valid_plus[col].isnull()
```

```python
#Imputation
my_imputer = SimpleImputer()
imputed_X_train_plus = pd.DataFrame(my_imputer.fit_transform(X_train_plus))
imputed_X_valid_plus = pd.DataFrame(my_imputer.fit_transform(X_valid_plus))
```

```python
imputed_X_train_plus.head()
```

```python
#Imputation removed column names we put them back
imputed_X_train_plus.columns = X_train_plus.columns
imputed_X_valid_plus.columns = X_valid_plus.columns
```

```python
MAE3 = score_dataset(imputed_X_train_plus, imputed_X_valid_plus, y_train, y_valid)
print("MAE from Approach 3 Imputation Extention : %d" % MAE3)
```

```python
#Approach 4 to drop the column with highest 0 values and imputation on rest.
```

```python
Xtrain1 = X_train.replace(0, np.nan)
Xvalid1 = X_valid.replace(0, np.nan)
```

```python
Xtrain1.isnull().sum()
```

```python
#Drop the column with highest null values
X_train_drop = Xtrain1.drop('2ndFlrSF', axis=1)
X_valid_drop = Xvalid1.drop('2ndFlrSF', axis=1)
```

```python
X_train_drop.head()
```

```python
my_2imputer = SimpleImputer()
imputed_X_train_drop = pd.DataFrame(my_2imputer.fit_transform(X_train_drop))
imputed_X_valid_drop = pd.DataFrame(my_2imputer.transform(X_valid_drop))
```

```python
imputed_X_train_drop.head()
```

```python
MAE4 = score_dataset(imputed_X_train_drop, imputed_X_valid_drop, y_train, y_valid)
print("MAE from Approach 4 drop column and impute:%d" % MAE4) 
```
